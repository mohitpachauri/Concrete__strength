```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns ; sns.set()
from matplotlib.pyplot import style ; style.use('ggplot')
```

```python
df = pd.read_csv('Concrete_Data_Yeh.csv')
df.head()
```

```python
plt.figure(figsize=(24,22))
sns.pairplot(df)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler as SS
ss = SS()
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
from sklearn.model_selection import train_test_split as tts
x_train , x_test , y_train , y_test = tts(x,y)
x_ss = ss.fit_transform(x,y)
xss_train , xss_test , yss_train , yss_test = tts(x_ss,y)
```

```python
from sklearn.decomposition import PCA
pca = PCA(2)
x_pca = pca.fit_transform(x,y)
```

```python
sns.scatterplot(x_pca[:,0],y)
```

```python
sns.scatterplot(x_pca[:,1],y)
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(2)
x_tsne = tsne.fit_transform(x,y)
```

```python
sns.scatterplot(x_tsne[:,0],y)
```

```python
sns.scatterplot(x_tsne[:,1],y)
```

```python
from sklearn.svm import SVR
svm = SVR(gamma=0.01)
```

```python
svm.fit(xss_train,yss_train)
svm.score(xss_test,yss_test)
```

```python
svm2 = SVR(gamma=0.01)
svm2.fit(x_train,y_train)
svm2.score(x_test,y_test)
```

```python
from sklearn.ensemble import ExtraTreesRegressor as ETR
etr = ETR()
etr2 = ETR()
etr.fit(x_train,y_train)
etr.score(x_test,y_test)
```

```python
etr2.fit(xss_train,yss_train)
etr2.score(xss_test,yss_test)
```

```python
y_pred = svm.predict(x_test)
```

```python
y_pred2 = etr2.predict(xss_test)
```

```python
from sklearn.ensemble import RandomForestRegressor as RFR
rf = RFR()
rf2 = RFR()
rf.fit(x_train,y_train)
rf2.fit(xss_train,yss_train)
```

```python
rf.score(x_test,y_test)
```
